```python
import pandas as pd
from sklearn.metrics import classification_report
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('cardio_train.csv', sep=';')
```

```python
df
```

```python
df.duplicated().sum()
```

```python
df = df.drop('id', axis=1)
```

```python
df
```

```python
df['age'] = (df['age'] / 365).astype('int32')
```

```python
df
```

```python
df['weight'] = df['weight'].astype('int32')
```

```python
df
```

```python
df.isna().sum()
```

```python
scaler = MinMaxScaler()
columns_to_scale = ['age', 'height', 'weight', 'ap_hi', 'ap_lo']
df[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])
```

```python
df
```

```python
X = df.drop('cardio', axis=1)
```

```python
y = df['cardio']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size=0.2, shuffle=True, random_state=42)
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
predictions = lr.predict(X_test)
```

```python
cr = classification_report(predictions, y_test)

print(cr)
```
